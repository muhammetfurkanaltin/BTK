```python
import matplotlib.pyplot as plt
import numpy as np 

x = [1,2,3,4]
y = [1.5,1.5,6,2]


plt.plot(x,y,"o--r")
plt.axis([0,6,0,20])

plt.title("Title")
plt.xlabel("x title")
plt.ylabel("y title")


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,2,100)
plt.plot(x,x, label="linear",color = "red")
plt.plot(x,x**2, label="quadratic",color="green")
plt.plot(x,x**3, label="cubic",color = "blue")

plt.title ("Simple Title")
plt.xlabel("x label")
plt.ylabel("y label")



plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,2,100)
fig,axs = plt.subplots(3)

axs[0].plot(x, x, color = "red")
axs[0].set_title("linear")

axs[1].plot(x, x**2, color="green")
axs[1].set_title("quad")

axs[2].plot(x, x**3, color="yellow")
axs[2].set_title("cubic")

plt.tight_layout()

```

```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,2,100)

fig,axs = plt.subplots(2,2)

fig.suptitle("title")

axs[0,0].plot(x, x, color = "yellow")
axs[0,1].plot(x, x**2, color = "red")
axs[1,0].plot(x, x**3, color = "blue")
axs[1,1].plot(x, x**4, color = "green")

plt.tight_layout()
```

# 21.3 Figure Oluşturma

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(-10,9,20)
y = x ** 3 
z = x ** 2

figure = plt.figure()
axes_cube = figure.add_axes([0.1,0.1,0.8,0.8])

axes_cube.plot(x,y,"b")
axes_cube.set_xlabel("X label")
axes_cube.set_ylabel("Y label")
axes_cube.set_title ("Cube")

axes_square = figure.add_axes([0.14,0.6,0.25,0.25])
axes_square.plot(x,z,"r")
axes_square.set_xlabel("X label")
axes_square.set_ylabel("Y label")
axes_square.set_title ("Square")

plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(-10,9,20)
y = x ** 3 
z = x ** 2
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])

axes.plot(x,z,label = "Saquare")
axes.plot(x,y,label = "Cube")

axes.legend(loc=4)

plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(-10,9,20)
y = x ** 3 
z = x ** 2
figure = plt.figure()

fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (4,4))
axes[0].plot(x,y)
axes[0].set_title("Cube")
axes[1].plot(x,z)
axes[1].set_title("Square")

plt.tight_layout()
fig.savefig("figure1.png")


plt.show()
```

# 21.4 Grafik Türleri - Pie Grafigi


```python
import matplotlib.pyplot as plt 


yil =[2011,2012,2013,2014,2015]
oyuncu1 = [8,10,12,7,9]
oyuncu2 = [7,12,5,15,21]
oyuncu3 = [18,20,12,25,19]
# Stack Plot

plt.plot([],[],color = "y",label = "oyuncu1")
plt.plot([],[],color = "r",label = "oyuncu2")
plt.plot([],[],color = "b",label = "oyuncu3")

plt.stackplot(yil,oyuncu1,oyuncu2,oyuncu3, colors =["y","r","b"])
plt.title("yıllara göre atılan goller")
plt.xlabel("Gol sayısı")
plt.ylabel("Yıllara göre")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt 


yil =[2011,2012,2013,2014,2015]
oyuncu1 = [8,10,12,7,9]
oyuncu2 = [7,12,5,15,21]
oyuncu3 = [18,20,12,25,19]
goal_types = "Penaltı", "Kaleye Atılan Sut", "Serbest Vrusu"

goals = [12,35,7]

colors = ["y","r","b"]

plt.pie(goals,labels = goal_types,colors=colors,shadow = True,explode =(0.05,0.05,0.05),autopct="%1.1f%%")

plt.show()
```
